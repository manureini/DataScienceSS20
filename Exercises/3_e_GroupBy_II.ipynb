# GroupBy II

```python
import pandas as pd
```

### Use Case: statistics on Occupations. 

```python
users = pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', 
                      sep='|', index_col='user_id')
users.head()
```

### Get the mean age per occupation

```python
users.groupby('occupation').age.mean()
```

### Get the Male ratio per occupation and sort it from the most to the least

```python
users['gender_number'] = users['gender'].apply(lambda r : 1 if r == 'M' else 0)
users
```

```python
groupedOccu = users.groupby('occupation');

a = groupedOccu.gender_number.sum() / groupedOccu.gender_number.count()
a.sort_values()
```

### For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation').age.agg(['min', 'max'])
```

### For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation', 'gender']).age.mean()
```

### [dificult] For each occupation present the percentage of women and men

```python
gender_ocup = users.groupby(['occupation', 'gender']).agg({'gender': 'count'})
 
total = gender_ocup.groupby('occupation').gender.sum()

result = pd.merge(gender_ocup, total, on='occupation', how='inner')

result["percent"] = result.gender_x / result.gender_y

result
```
